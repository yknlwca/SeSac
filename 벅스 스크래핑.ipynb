```python
driver = webdriver.Chrome()
url = "https://music.bugs.co.kr/chart/track/realtime/total?wl_ref=M_contents_03_01"
driver.get(url)
html = driver.page_source
soup = BeautifulSoup(html,"html.parser")
bucks_list = soup.select("tr")[1:-3]
Bucks_list = []
rank = 1
for idx in bucks_list:
    song_title = idx.select(".title")[0].text.strip()
    Artist = idx.select(".artist")[0].text.strip()
    Album = idx.select("td.left > a")[0].text.strip()
    Bucks_list.append([rank,song_title,Artist,Album])
    rank+=1
df = pd.DataFrame(Bucks_list, columns = ["순위","제목","가수","앨범"])
df = df.set_index("순위")
df.to_excel("Bucks_list.xlsx",index=False)
driver.quit()
```
