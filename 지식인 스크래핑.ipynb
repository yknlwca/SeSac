```python
question_list = []
for i in range(1,42,10):
    url = "https://search.naver.com/search.naver?where=kin&kin_display=10&qt=&title=0&&answer=0&grade=0&choice=0&sec=0&nso=so%3A-1%2Ca%3A%2Cp%3Aall&query=%EB%94%A5%EB%9F%AC%EB%8B%9D&c_id=&c_name=&sm=tab_pge&kin_start="+str(i)+"&kin_age=0"
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html,"html.parser")
    for idx in range(1,11):
        question = soup.select(f"li:nth-child({idx}) > div > .question_area > .question_group> a")[0].text.strip()
        question_list.append([question])
        question_list
    print(question_list)
```
