```python
data[['computerUsageTime', 'tvUsageTime', 'airconWithHeatUsageTime',
       'airconUsageTime', 'heaterUsageTime', 'elecStoveUsageTime',
       'elecBlanketUsageTime', 'airFryerUsageTimePerWeek']]
```

```python
def process_row(row):
    numbers = list(map(float, row.split(',')))
    return numbers[0] * 5 + numbers[1] * 2

data['new_computerUsageTime'] = data['computerUsageTime'].apply(process_row)
data['new_tvUsageTime'] = data['tvUsageTime'].apply(process_row)
data['new_airconUsageTime'] = data['airconUsageTime'].apply(process_row)
data['new_heaterUsageTime'] = data['heaterUsageTime'].apply(process_row)
data['new_elecStoveUsageTime'] = data['elecStoveUsageTime'].apply(process_row)
data['new_elecBlanketUsageTime'] = data['elecBlanketUsageTime'].apply(process_row)
```

```python
data_month=data[[ 'electricalAppliances', 'month1', 'month2',
       'month3', 'month4', 'month5', 'month6', 'month7', 'month8', 'month9',
       'month10', 'month11', 'month12']]
data_month
```

```python
import pandas as pd

for i in range(1, 31):
    data_month[f'appliance_{i}'] = 0

# Function to safely convert to set of integers, handling empty and non-numeric values
def safe_int_set(s):
    try:
        return set(map(int, s.split(', ')))
    except ValueError:
        return set()

# Function to process each row
def process_row(row):
    appliances = safe_int_set(row['electricalAppliances'])
    for month_col in ['month1', 'month2', 'month3', 'month4', 'month5', 'month6', 
                      'month7', 'month8', 'month9', 'month10', 'month11', 'month12']:
        if pd.notna(row[month_col]):  
            month_usage = safe_int_set(row[month_col])
            for appliance in appliances:
                if appliance in month_usage:
                    row[f'appliance_{appliance}'] += 1
    return row


data_month = data_month.apply(process_row, axis=1)

data_month

```
